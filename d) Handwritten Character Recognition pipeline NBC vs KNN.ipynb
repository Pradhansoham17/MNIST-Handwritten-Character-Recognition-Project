```python
import numpy as np
import pandas as pd
train_df=pd.read_csv("training-images.csv",header=None)
train_df.rename(columns={ train_df.columns[784]: "Label" }, inplace = True)
train_df
```

```python
test_df=pd.read_csv("test-images.csv",header=None)
test_df.rename(columns={ test_df.columns[784]: "Label" }, inplace = True)
test_df
```

```python
comb_df=pd.concat([train_df, test_df], axis=0, ignore_index=True)
comb_df
```

```python
#Site :- "https://chrisalbon.com/machine_learning/model_selection/model_selection_using_grid_search/"
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

# Create a pipeline
pipe = Pipeline([('classifier', KNeighborsClassifier())])

# Create space of candidate learning algorithms and their hyperparameters
search_space = [{'classifier': [GaussianNB()]},
                {'classifier': [KNeighborsClassifier()],
                 'classifier__n_neighbors': [i for i in range(1,11)]}]

# Create grid search 
clf = GridSearchCV(pipe, search_space, cv=5, verbose=10)
clf
```

```python
X = comb_df.iloc[:,:]
y = comb_df.iloc[:,[-1]]

# Fit grid search
clf.fit(X, y)
```

```python
cv_results_=clf.cv_results_
cv_results_
```

```python
dfcv = pd.DataFrame(data=cv_results_)
dfcv
```
