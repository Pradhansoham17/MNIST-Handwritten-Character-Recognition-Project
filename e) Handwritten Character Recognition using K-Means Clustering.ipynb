```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
train_df=pd.read_csv("training-images.csv",header=None)
train_df.rename(columns={ train_df.columns[784]: "Label" }, inplace = True)
train_df
```

```python
test_df=pd.read_csv("test-images.csv",header=None)
test_df.rename(columns={ test_df.columns[784]: "Label" }, inplace = True)
test_df
```

```python
comb_df=pd.concat([train_df, test_df], axis=0, ignore_index=True)
comb_df
```

```python
from sklearn.cluster import KMeans
import numpy as np
X = comb_df.iloc[:,:-1]
kmeans = KMeans(n_clusters=4, random_state=5).fit(X)
```

```python
kmeans.labels_
```

```python
temp_df=comb_df
temp_df["Cluster"]=kmeans.labels_
temp_df
```

```python
kmeans.predict([X.iloc[0,:]]) #predicting for first row
```

```python
#final cluster centers after algorithm is completed
kmeans.cluster_centers_
```

```python
#Sum of squared distances of samples to their closest final cluster center
kmeans.inertia_ 
```
