```python
import numpy as np
import pandas as pd
train_df=pd.read_csv("training-images.csv",header=None)
train_df.rename(columns={ train_df.columns[784]: "Label" }, inplace = True)
train_df
```

```python
test_df=pd.read_csv("test-images.csv",header=None)
test_df.rename(columns={ test_df.columns[784]: "Label" }, inplace = True)
test_df
```

```python
X_train=train_df.sample(frac=1, random_state=5).iloc[:,:-1]
y_train=train_df.sample(frac=1, random_state=5).iloc[:,[-1]]
```

```python
X_test=test_df.sample(frac=1, random_state=5).iloc[:,:-1]
y_test=test_df.sample(frac=1, random_state=5).iloc[:,[-1]]
```

```python
X = X_train.iloc[:,:]
y = y_train['Label']
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
```

```python
neigh.classes_
```

```python
neigh.predict_proba(X_test.iloc[[0]])
```
