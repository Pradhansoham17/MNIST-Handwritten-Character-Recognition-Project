```python
import numpy as np
import pandas as pd
train_df=pd.read_csv("training-images.csv",header=None)
train_df.rename(columns={ train_df.columns[784]: "Label" }, inplace = True)
train_df
```

```python
test_df=pd.read_csv("test-images.csv",header=None)
test_df.rename(columns={ test_df.columns[784]: "Label" }, inplace = True)
test_df
```

```python
X_train=train_df.sample(frac=1, random_state=5).iloc[:,:-1]
y_train=train_df.sample(frac=1, random_state=5).iloc[:,[-1]]
```

```python
X_test=test_df.sample(frac=1, random_state=5).iloc[:,:-1]
y_test=test_df.sample(frac=1, random_state=5).iloc[:,[-1]]
```

```python
import numpy as np
X = np.array(X_train.values)
y = np.array(y_train.values)
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB() 
#clf = GaussianNB(priors=[0.5, 0.5]) #here priors are set to 50%, remove for improving accuracy
#priors is the distributioon of output classes in the data
clf.fit(X, y)
```

```python
# first row of our test data
a=np.array(X_test.values)[0]
a
```

```python
#to check the prediction of a single row
#checking the prediction of our first row in test data
predictSingle=clf.predict([a]) #we enclose 'a' in '[]' because we are supposed to pass a 2-D array to the .predict() function
#so we take our first row, which is a 1-D array and convert it to a 2-D array by enclosing it in '[]' 
print(predictSingle)
print(type(predictSingle))
```

```python
clf.classes_
```

```python
predlist=clf.predict(np.array(X_test.values))
print(predlist)
print(type(predlist))
```

```python
#conditional probability outcome for [0, 1, 2, 3] of each row of test data
proba_list=clf.predict_proba(np.array(X_test.values))
proba_list
```

```python
clf.predict_proba([np.array(X_test.values)[0]])
```

```python
# score method in sklearn GaussianNB
accuracy=clf.score(X_test,y_test['Label'])
print('The Accuracy of algorithm is : ',accuracy*100, '%')
```
